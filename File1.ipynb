# Plotting graphs

### Plotting a graph of y = sqrt(x)

````python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 100, 400)
y = np.sqrt(x)
plt.plot(x, y, label='y = $\sqrt{x}$', color='red')
plt.legend() # Shows the labels for Sine and Cosine

plt.show()
````

### Graph showing average temperature across certain months 

````python
import matplotlib.pyplot as plt

temp = [27, 30, 33, 34, 33, 29]
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']

plt.plot(months,temp, label = 'title')
plt.title("Average temp. vs Months")
plt.show()

````
